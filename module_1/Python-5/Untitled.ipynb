```python
#Определите, во сколько раз средняя зарплата (Wage) голкипера (Position, GK) с максимальным значением показателя " Рефлексы" 
#(GKReflexes) 
#выше средней зарплаты голкипера с максимальным значением показателя "Владение мячом" (GKHandling). Ответ округлите до сотых.
```

```python
(football[football.GKReflexes.max() == football.GKReflexes].Wage.mean())/(football[football.GKHandling.max() == football.GKHandling].Wage.mean())
```

```python
#Определите, во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков 
#(игроков с максимальным значением показателя "Агрессивность" (Aggression)) 
#выше средней силы удара игроков с минимальной агрессией. Ответ округлите до сотых.

(football[football.Aggression.max() == football.Aggression].ShotPower.mean())/(football[football.Aggression.min() == football.Aggression].ShotPower.mean())
```
