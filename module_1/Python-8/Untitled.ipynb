```python
log = pd.read_csv("log.csv",header=None) 
log.columns = ['user_id','time','bet','win']

def something(x):
    if 'user' in x:
        return x.replace('Запись пользователя № - ','')

def something3(x):
    if '[' in str(x):
        return x.replace('[','')
        
log = log.drop(log[log.user_id.str.contains('error', na=False)].index)

log['user_id'] = log.user_id.apply(something)
log['time'] = log.time.apply(something3)
log
```
