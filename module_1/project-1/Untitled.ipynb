```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
data = pd.read_csv('movie_bd_v5.csv')
data.sample(5)
```

```python
data['cast'] = data['cast'].str.split('|')
```

```python
len(data.explode('cast')['cast'].unique())
```

```python
display(data)
```

```python
all_cast = (data.explode('cast')['cast'].unique())
intersections = dict.fromkeys(all_cast, None) #собираем словарь с именами актёров
```

```python
len(intersections)
```

```python
for _ in data['cast']: 
    for actor in _:
        for actor2 in _:
            if actor != actor2:
                if not intersections[actor]:
                    intersections[actor] = {}
                try:
                    intersections[actor][actor2] += 1
                except KeyError:
                    intersections[actor][actor2] = 1
```
