```python
import pandas as pd
import json
from pprint import pprint
```

```python
with open('recipes.json') as f:
    recipes = json.load(f)
```

```python
ing_set = set()

for recipe in recipes:
    for items in recipe['ingredients']:
        ing_set.add(items)
        #ing_set.add(items['ingredients'])
len(ing_set)       
```

```python
df = pd.DataFrame(recipes)
df
```

```python
def find_item(cell):
    if item in cell:
        return 1
    return 0
```

```python
for item in ing_set:
    df[item] = df['ingredients'].apply(find_item)
```

```python
df['ingredients'] = df['ingredients'].apply(len)
```
